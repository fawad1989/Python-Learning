Forward or Ascending Indexing starts from 0 and go to the len(array) - 1

Reverse or Descending Indexing starts from -1 to -(len(array))

The requirement is first five elements then the slice would be array[0:5] where the 5 is exluded as first 5 values are at at index 0,1,2,3,4.

# Combining Arrays

Arrays can be combined in many ways. In Numpy it is called stacking. It can be horizontal, vertical and even depth waise.

```python
a = np.arange(9).reshape(3,3)
b = np.arange(0,-9,-1).reshape(3,3)
print('a:', a)
print('-------------------')
print('b:', b)
```

```python
# only stack will put the first array atop the other

np.stack((a,b))
```

# Horizontal stacking

will stack the array beside the other array as columns. One important thing is that the number of rows should be the same otherwise there will be error

```python
a = np.arange(16).reshape(8,2)
b = np.arange(16,56).reshape(8,5)
print('a:', a)
print('-------------------')
print('b:', b)
```

```python
np.hstack((a,b))
```

```python
c = b.reshape(40,1) # reshaped bo to a new shape where rows are different from the other one
```

```python
np.hstack((a,c)) # error
```

```python
# other method for horizontal stack

np.concatenate((a,b),axis=1) #axis = 1 tells the operation to do horizontal stacking
```

# Vertical stacking

will stack the array beside the other array as rows. One important thing is that the number of columns should be the same otherwise there will be error

```python
a = np.arange(20).reshape(2,10)
b = np.arange(20,60).reshape(4,10)
print('a:', a)
print('-------------------')
print('b:', b)
```

```python
np.vstack((a,b))
```

```python
np.concatenate((a,b),axis = 0)
```

# Depth Stacking

It takes two arrays and stacks them depth wise. It takes each column of array and stack then in the third axis

```python
a = np.arange(9).reshape(3,3)
b = np.arange(0,-9,-1).reshape(3,3)
print('a:', a)
print('-------------------')
print('b:', b)
```

```python
c = np.dstack((a,b))
print('C:\n',c)
print(c.shape)
```

```python
# OR
#axis wise stacking

c_0 = np.stack((a,b), axis = 0) # by default axis = 0 but just to understand its written
c_1 = np.stack((a,b), axis = 1) # axis = 1 ROW wise depth will be performed
c_2 = np.stack((a,b), axis = 2) # axis = 1 Column wise depth will be done
```

# Numerical Methods included in Numpy

```python
a = np.arange(9).reshape(3,3)
a
```

```python
min_val= np.min(a)
max_val= np.max(a)
min_val_pos= np.argmin(a)
max_val_pos= np.argmax(a)
print('Min value of array a:', min_val)
print('Max value of array a:', max_val)
print('Pos of min value in array a:', min_val_pos)
print('Pos of max value in array a:', max_val_pos)
# Also to find the index
loc_min = np.where(a == min_val) # it will give the index of the min value in a
loc_max = np.where(a == max_val) # it will give the index of the max value in a
print('Index of min value:', loc_min)
print('Index of max value:', loc_max)
# Row or Column Wise Operation
min_val_pos_col= np.min(a, axis = 0) # Col operation
max_val_pos_col= np.max(a, axis = 0) # Col operation
min_val_pos_row= np.min(a, axis = 1) # row operation
max_val_pos_row= np.max(a, axis = 1) # row operation
print('min value in array a Col wise:', min_val_pos_col)
print('max value in array a Col wise:', max_val_pos_col)
print('min value in array a row wise:', min_val_pos_row)
print('max value in array a row wise:', max_val_pos_row)


```

# Matrix Multiplication and Division.

$ [C] = [A] * [B]$ Point wise or element wise 


$ [C] = [A] * B$ Scalar Multiplication

$ [C] = [A] .dot [B]$ Matrix Multiplication OR 

$<a,b> = a^T b = \sum_{i=1}^n a_i b_i$

$ [C] = [A] / [B]$ Porintwise or element wise division




```python
# Multiplication

# Point or element wise Multiplication
D = a*b # D= np.multiply(a,b, dtype = np.int32)
print(D)
print('--------------')

# Matrix multiplication is different

D = a.dot(b) # or
# D = np.matmul(a,b) # or 
# D = a@b 
print(a)
print(b)
print(D)
```

```python
# inner Product of two arrays

np.inner(a,b) # row wise operation
```

# Cholesky Decomposition: 
$ A = UU^T$

```python
a = np.array([[4,12,-16],
              [12,37,-43],
              [-16,-43,98]])
print(a)
print('------Cholesky Decomposition------')
cd = np.linalg.cholesky(a)
print(cd)
print('----------Verification------------')
a = np.matmul(cd,cd.transpose())
print(a)
```

```python
# Complex Matrix
a = np.array([[1,-2j],
              [2j,5]])
print(a)
print('--------Cholesky Decomposition---------')
cd = np.linalg.cholesky(a)
print(cd)
print('--------------Verfication--------------')
cdv = np.matmul(cd,cd.T.conj())
print(cdv)
```

# QR Decomposition
$ A = QR$

returns two output i.e. Q and R

```python
a = np.array([[4,12,-16],
              [12,37,-43],
              [-16,-43,98]])
print(a)
```

```python
qrd = np.linalg.qr(a)
print('Q=',qrd[0])
print('------------------')
print('R=',qrd[1])
print('-------------Verification-----------')
av = np.matmul(qrd[0],qrd[1])
print(av)
```

# Eigen Value Decomposition
$ A = PDP^{-1}$

where A is Eigen Vectors

D is Diagonal Matrix consisting Eigen values of A 

P is a sqaure matrix


```python
a = np.array([[1,2],[0,3]])
print(a)
(D,P) = np.linalg.eig(a) #first output is D and other output is P
print('D:',np.diag(D)) # origianl output is a 1D array
print('P:',P)
print('---------Verfication------------')
av = np.matmul(np.matmul(P,np.diag(D)),np.linalg.inv(P))
print(av)
```

# Matrix Norms

FUnction that take matric as input and gives a number as output

Frobenius
$ \|A\|=\sqrt{\sum_{i=1}^m\sum_{j=1}^n\lvert a_{ij}\rvert^{2}}$

Determinant for 2x2 Matrix

$A = \begin{bmatrix}
a & b\\
c & d 
\end{bmatrix} $

$ \lvert A\rvert\ = ad - bc$

Determinant for 3x3 Matrix
$A = \begin{bmatrix}
a & b & c\\
d & e & f\\
g & h & i
\end{bmatrix} $

$\lvert A\rvert\ = a(ei - fh) - b(di - fg) + c(dh - eg)$


```python
# Frobenius Norm

a = np.arange(1,10).reshape(3,3)
print(a)
print('-------Frobenius-----------')
afn = np.linalg.norm(a)
print(afn)
```

```python
# Conditional Norm tells you how close the matrix is to singularity
a = np.arange(1,10).reshape(3,3)
print(a)
print('-------Conditional Number-----------')
cdn = np.linalg.cond(a)
print(cdn)
```

```python
# Determinant of Matrix

a = np.array([[3,8],[4,6]])
b = np.array([[6,1,1],[4,-2,5],[2,8,7]])

adet = np.linalg.det(a)
bdet = np.linalg.det(b)

print(a)
print(b)
print('-----------Determinant----------')
print(adet)
print(bdet)
```

# Solving Matrix Equation

$ Ax = B$

Find x

```python
A = np.array([[1,5,6],[1,8,9],[0,-1,6]])
B = np.random.random((3,1))

print(A)
print(B)
```

```python
x = np.matmul(np.linalg.inv(A),B)
print(x)
# OR

print('---------Solve----------')

x = np.linalg.solve(A,B)
print(x)
```
